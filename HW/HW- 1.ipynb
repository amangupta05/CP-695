```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python

plt.style.use('seaborn-whitegrid') 
import numpy as np 
noise_scale = 100 
number_of_samples = 50 
x =  30*(np.random.rand(number_of_samples, 2) - 0.5) 
y = 5 * x + 15 * x**2 + 2 * x*3+ noise_scale*np.random.randn(number_of_samples, 1) 
plt.plot(x,y,'ro') 
```

Task 1

```python
plt.style.use('seaborn-whitegrid') 
m=4
X=30*(np.random.rand(number_of_samples, 1) - 0.5)
Y= 5 * X + 15 * X**2 + 2 * X*3+ noise_scale*np.random.randn(number_of_samples, 1) 
poly=PolynomialFeatures(degree=m,include_bias=False)
poly_X=poly.fit_transform(X)
plt.plot(poly_X,Y,'ro')
```

Task 2


```python
def MSE(X,Y):
    m= [1, 2, 3, 4, 5, 6, 7, 8]
    n=len(X)
    MSE=[]
    for i in m:
        poly=PolynomialFeatures(degree=i,include_bias=False)
        Poly_X=poly.fit_transform(X)
        model=LinearRegression()
        model.fit(Poly_X,Y)
        y_pred=model.predict(Poly_X)
        mse=mean_squared_error(Y,y_pred)
        MSE.append(mse)
    return MSE
        
        
```

```python
mse=MSE(X,Y)
m= [1, 2, 3, 4, 5, 6, 7, 8]
```

```python

plt.plot(m,mse)
plt.xlabel('m')
plt.ylabel('MSE')
plt.show()
```

For m= 2 we get lowest MSE

```python
noise=[200, 300, 400, 600, 800, 1000]

for j in range(len(noise)):
     
    X3=30*(np.random.rand(number_of_samples, 1) - 0.5)
    y3=5 * X3 + 15 * X3**2 + 2 * X3*3+ noise[j]*np.random.randn(number_of_samples, 1)
    best_poly=PolynomialFeatures(degree=2,include_bias=False)
    best_poly_X= best_poly.fit_transform(X3)
    plt.figure
    plt.style.use('seaborn-whitegrid')
    plt.title('noise:'+ str(noise[j]) )
    plt.xlabel('polynomial of degree 2')
    plt.ylabel('data with noise')
    
    plt.plot(best_poly_X,y3,'ro')
    plt.show()
```
